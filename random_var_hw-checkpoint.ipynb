```python
import numpy as np
import pandas as pd
import random
successes = 0    # Количество успехов
lose=0
trials=100
startmoney=1 #количество денег на начало игры
for p in range(100):
        number = random.randint(0,999)
        if number==777:
            startmoney+=200
            successes += 1
        elif number==999:
            startmoney+=100
            successes+=1
        elif number==555:
            startmoney+=50
            successes+=1
        elif number==333:
            startmoney+=15
            successes+=1
        elif number==111:
            startmoney+=10
            successes+=1
        elif number%100==77:
            startmoney+=5
            successes+=1
        elif number%100==7:
            startmoney+=3
        elif number%100==0 and number%10==0:
            startmoney+=2
            successes+=1
        elif number%100==0:
            startmoney+=1
            successes+=1
        else:
            lose+=1
          #  startmoney=startmoney-1
        p_monte_carlo = successes/trials #вероятность выигрыша
        lst_avg = np.average(startmoney)
        print(successes,startmoney,lose,p_monte_carlo,lst_avg)
        
        
```
