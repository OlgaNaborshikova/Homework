```python
#работа с пропусками
print('------------------------')
#surgery? можно заменить медианой, так как данные при сравнении статистик в целом не отличается
new_df = pd.read_csv('horse_data.csv',names=col_names)
new_df['surgery?'].fillna(df['surgery?'].median(), inplace=True)
print(new_df['surgery?'].describe())
print(df['surgery?'].describe())
print('------------------------')
#Age не требует работы с пропусками, так как подсчет пропуском показал -0
#rectal temperature, тоже можно заменить медианой
new_df = pd.read_csv('horse_data.csv',names=col_names)
new_df['rectal temperature'].fillna(df['rectal temperature'].median(), inplace=True)
print(new_df['rectal temperature'].describe())
print(df['rectal temperature'].describe())
print('------------------------')
#pulse, замена медианой оптимальная
new_df = pd.read_csv('horse_data.csv',names=col_names)
new_df['pulse'].fillna(df['pulse'].median(), inplace=True)
print(new_df['pulse'].describe())
print(df['pulse'].describe())
print('------------------------')
#выбрала медиану при зпмене данных respiratory rate, так как на мой взгляд, наиболее оптим.
new_df = pd.read_csv('horse_data.csv',names=col_names)
new_df['respiratory rate'].fillna(df['respiratory rate'].median(), inplace=True)
print(new_df['respiratory rate'].describe())
print(df['respiratory rate'].describe())
print('------------------------')
#temperature of extremities заменяем медианой, так как стат. данные остаются приближенными
new_df = pd.read_csv('horse_data.csv',names=col_names)
new_df['temperature of extremities'].fillna(df['temperature of extremities'].median(), inplace=True)
print(new_df['temperature of extremities'].describe())
print(df['temperature of extremities'].describe())
#исходя из описания столбца-pain, чем выше его показатель, тем большая вероятность что потребуется хирург. вмешательство. Поэтому так же стоит заполнить медианой, и основния по хирургии делать не только на одном столбце
new_df = pd.read_csv('horse_data.csv',names=col_names)
new_df['pain'].fillna(df['pain'].median(), inplace=True)
print(new_df['pain'].describe())
print(df['pain'].describe())
#outcom тое бы заменила медианой, так как по статистикам видим, что большее количество лошадей,живы
new_df = pd.read_csv('horse_data.csv', names=col_names)
new_df['outcome'].fillna(df['outcome'].median(), inplace=True)
print(new_df['outcome'].describe())
print(df['outcome'].describe())
#Плоэтому изучим имеющиеся данные можно применит медиану ко всем пропускам
print('вывод датафрейма без пропусков')
df.iloc[:, [0, 1, 3, 4, 5,6, 10, 22]].fillna(df.iloc[:, [0, 1, 3, 4, 5,6, 10, 22]].median())
```
