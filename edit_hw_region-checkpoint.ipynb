```python
movies = pd.read_csv('movies.csv') 
ratings=pd.read_csv('ratings.csv')
joined=ratings.merge(movies,on='movieId',how='left')
class_row = joined.pivot_table(index=['title','movieId'],values='rating',aggfunc='mean')
class_row.head()
def classifficaction(rating):
    if rating <= 2:   
            return ('низкий рейтинг')
    elif rating <= 4 and rating > 2:
            return ('средний рейтинг')
    else:
            return ('высокий рейтинг')
class_row['class'] = class_row['rating'].apply(classifficaction)
class_row.head()
```
